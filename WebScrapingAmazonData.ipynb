```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.amazon.in/s?k=pants+women&crid=1PC0RPVO2OMOM&sprefix=pants+wo%2Caps%2C222&ref=nb_sb_ss_ts-doa-p_1_8'
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage = requests.get(url, headers = headers)
webpage
```

```python
soup = BeautifulSoup(webpage.content, 'html.parser')
soup
```

```python
element = soup.find_all('div', class_ = 'a-section a-spacing-base a-text-center')
len(element)
```

```python
names = []
brand = []
ratings = []
sellingPrice = []
actualPrice = []
for i in element:
    names.append(i.find('span', class_ = 'a-size-base-plus a-color-base a-text-normal').text.strip())
    brand.append(i.find('span', class_ = 'a-size-base-plus a-color-base').text.strip())
    ratings.append(i.find('span', class_ = 'a-icon-alt'))
    sellingPrice.append(i.find('span', class_ = 'a-price-whole').text.strip())
    actualPrice.append(i.find('span', class_ = 'a-offscreen').text.strip())

d = {'name':names, 'brand':brand, 'rating':ratings, 'Selling Price':sellingPrice, 'Actual Price':actualPrice}
df = pd.DataFrame(d)
df
```

```python
final = pd.DataFrame()
for j in range(1, 8):
    url = 'https://www.amazon.in/s?k=pants+women&crid=1PC0RPVO2OMOM&qid=1717862490&sprefix=pants+wo%2Caps%2C222&ref=sr_pg_{}'.format(j)
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get(url, headers = headers)
    soup = BeautifulSoup(webpage.content, 'html.parser')
    element = soup.find_all('div', class_ = 'a-section a-spacing-base a-text-center')
    names = []
    brand = []
    ratings = []
    sellingPrice = []
    actualPrice = []
    for i in element:
        names.append(i.find('span', class_ = 'a-size-base-plus a-color-base a-text-normal').text.strip())
        brand.append(i.find('span', class_ = 'a-size-base-plus a-color-base').text.strip())
        ratings.append(i.find('span', class_ = 'a-icon-alt'))
        sellingPrice.append(i.find('span', class_ = 'a-price-whole').text.strip())
        actualPrice.append(i.find('span', class_ = 'a-offscreen').text.strip())

    d = {'name':names, 'brand':brand, 'rating':ratings, 'Selling Price':sellingPrice, 'Actual Price':actualPrice}
    df = pd.DataFrame(d)
    final = pd.concat([final, df], ignore_index = True)

print(final)
```

```python
file_name = 'Amazon_pants_women.csv'
final.to_csv(file_name, index = False)
```
